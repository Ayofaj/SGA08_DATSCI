```python
import pandas as pd 
import matplotlib.pyplot as plt 
import numpy as np 
bicycle = pd.read_csv('train.csv')
```

```python
bicycle.weather.unique()
```

```python
bicycle.groupby('weather')['casual'].mean().to_frame(name = 'casual').reset_index()
```

```python
bicycle.groupby('weather')['casual'].mean().plot(kind='bar')
```

```python
bicycle.groupby('weather')['registered'].mean().to_frame(name = 'registered').reset_index()
```

```python
bicycle.groupby('weather')['registered'].mean().plot(kind='bar')
```
